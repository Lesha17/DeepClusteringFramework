```python
import torch
from clustering_tool.datasets import *
from clustering_tool.train import *
from clustering_tool.clusterer import *
from clustering_tool.model import *
from clustering_tool.metrics import *
from clustering_tool.classify import *
```

```python
device='cpu'
if torch.cuda.is_available():
    device='cuda'
```

```python
def train_model(model, dataloader, loss_fn, with_decoder=False):
    losses={}
    if not with_decoder:
        model.decoder = None
        losses = {'clusterer_loss': constant_loss_weight_fn(1.0)}
    else:
        decoder_weight = 0.5
        losses = {'clusterer_loss': constant_loss_weight_fn(1 - decoder_weight), 
                  'decoder_loss': constant_loss_weight_fn(decoder_weight)}
        
    start_lr = 1e-4
    end_lr = 1e-5
    num_epochs = 2
    gamma = (end_lr / start_lr) ** (1/num_epochs)
    train(model, dataloader, losses, start_lr, num_epochs=num_epochs, gamma=gamma, progress_each_iteration=False)
    
def save_metrics(model, dataloader, data_name, emb_name, loss_name):
    dec=''
    if model.decoder is not None:
        dec = 'dec'
    with open('metrics/{}_{}_{}{}.json'.format(data_name, emb_name, loss_name, dec), 'w') as metric_file:
        metric_file.write(str(calculate_metrics(model, dataloader)) + '\n')
        metric_file.write(str(classify(model, dataloader)))
```

```python
datasets = ['SearchSnippets', 'Biomedical', 'StackOverflow']
emb_types = ['word2vec_md_avg', 'word2vec_lg_avg', 'word2vec_md_weighted','word2vec_lg_weighted', 
             'bert_avg', 'bert_cls', 'bert_pooler', 'bert_pooler']
losses = {'kl_div': kl_div_loss, 'dot_product': dot_product_loss, 'cross_entropy': cross_entropy_loss, 'mse': torch.nn.MSELoss()}
for data_name in datasets:
    for emb_name in emb_types:
        dataloader = read_saved('data/{}_{}.npy'.format(data_name, emb_name), 
                                'data/{}_label.npy'.format(data_name), device=device)
        model = torch.load('models/{}_{}_ae.pt'.format(data_name, emb_name))
        model.decoder=None
        save_metrics(model, dataloader, data_name, emb_name, 'initial')
        for loss_name, loss_fn in losses.items():
            model = torch.load('models/{}_{}_ae.pt'.format(data_name, emb_name))
            model.clusterer.loss_fn = loss_fn
            train_model(model, dataloader, loss_fn, with_decoder=False)
            torch.save(model, 'models/results/{}_{}_{}.pt'.format(data_name, emb_name, loss_name))
            save_metrics(model, dataloader, data_name, emb_name, loss_name)
            
            model = torch.load('models/{}_{}_ae.pt'.format(data_name, emb_name))
            model.clusterer.loss_fn = loss_fn
            train_model(model, dataloader, loss_fn, with_decoder=True)
            torch.save(model, 'models/results/{}_{}_{}_dec.pt'.format(data_name, emb_name, loss_name))
            save_metrics(model, dataloader, data_name, emb_name, loss_name)
            
```
